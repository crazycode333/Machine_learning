```python
my_data=np.random.randn(3,4)
my_data
```

```python
df=pd.DataFrame(data=my_data,index=['R1','R2','R3'],columns=['C1','C2','C3','C4'])
df
```

```python
#selecting specific columns
df[['C1','C2']]
```

```python
#add column C5 which is sum of C1 and C2
df['C5']=df['C1']+df['C2']
df
```

```python
#to delete a column C3 , axis=1 means column wise operation,axis=0 means row wise operation, inplace=True means changes will be made in the original dataframe
df.drop('C3',axis=1,inplace=False)
```

```python
df
```

```python
#Accessing rows by loc and iloc
#loc is label based indexing
#iloc is integer based indexing
df.loc['R1']
```

```python
df.iloc[0]
#df.iloc[-1]
```

```python
#selecting multiple rows and columns , matrix form
df.loc[['R1','R3'],['C4','C5']]
```

```python
#For example if we want to select all rows where value of column C3 is less than 0 and value of column C5 is greater than 0
#Multiple conditions should be enclosed in parentheses (AND is & , OR is |)
df[ (df['C3']<0) & (df['C5']>0 ) ]
```

```python
df[ (df['C3']<0) | (df['C5']>0 ) ]
```

```python
myindex=[('slot1',1),('slot1',2),('slot1',3),('slot2',1),('slot2',2),('slot2',3),('slot3',1),('slot3',2),('slot3',3)]
myindex=pd.MultiIndex.from_tuples(myindex)
dataframe=pd.DataFrame(data=np.random.randn(9,2),index=myindex,columns=['A','B'])
dataframe
```

```python
print(dataframe.loc['slot2'])  #accessing all data for slot2
print("----")
print(dataframe.loc['slot2'].loc[2])  #accessing data for slot2 and 2nd sub index
print('-----')
print(dataframe.xs('slot2') ) #accessing all data for slot2 using xs cross section method
print('-----')
print("Now i want to give name to the indexes for slots and sub indexes")
dataframe.index.names=['Slots','Subslots']
print(dataframe)

print(dataframe.xs(2,level='Subslots'))  #accessing all data for subslot 2 using xs cross section method
```
