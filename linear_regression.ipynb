```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('Housing.csv')
data.shape
```

```python
data.info()
```

```python
data.head()
```

```python
X=data.iloc[:,0].values
y=data.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train
```

```python
#Feature scaling is usually not required for linear regression as the algorithm can handle features of different scales.
# However, in some cases, feature scaling can help improve the convergence speed of the optimization algorithm used to train the model.
# If the features have very different scales, it can lead to numerical instability during the optimization process, which can slow down convergence
# we have to reshape X_train and X_test because they are 1D arrays and sklearn expects 2D arrays for the features.
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train.reshape(-1,1),y_train)
```

```python
#creating predictions
y_predict=regressor.predict(X_test.reshape(-1,1))
```

```python
y_predict
```

```python
#plotting info for training set
plt.scatter(X_train,y_train,)
plt.plot(X_train,regressor.predict(X_train.reshape(-1,1)))
plt.title('House Price vs Area (Training set)')
```

```python
#ploting info for test set
plt.scatter(X_test,y_test)
plt.plot(X_train,regressor.predict(X_train.reshape(-1,1)))
plt.title('House Price vs Area (Test set)')
```
