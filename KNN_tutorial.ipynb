```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline 
```

```python
data=pd.read_csv('Social_Network_Ads.csv')
```

```python
data.head(10)
```

```python
print(data.info())
print("***********************"*5)
print(data.isnull().sum())
```

```python
X=data.iloc[:,2:4].values
y=data.iloc[:,-1].values
```

```python
print(X)
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
print(X_train)
print(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=5,p=2,metric='minkowski')
classifier.fit(X_train,y_train)
```

```python
y_predict=classifier.predict(X_test)
```

```python
y_predict
```

#Confussion Matrix
A confusion matrix is a key tool in machine learning for evaluating the performance of classification models. It shows how well your model is predicting classes by comparing actual vs predicted labels.


A contour plot is a graphical representation used to show 3D data in two dimensions using contour lines. These lines connect points of equal value, making it easier to visualize gradients, peaks, and valleys in data.
📊 When to Use a Contour Plot
- To visualize functions of two variables (e.g., z=f(x,y))
- To analyze topography, temperature distributions, or loss landscapes in ML
- To compare regions of high and low intensity in a dataset
🧠 Key Components
- X and Y axes: Represent the two independent variables
- Contour lines: Connect points with equal Z values
- Color gradients: Often used to enhance visual interpretation

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

X, Y = np.meshgrid(x, y)

Z = np.sin(np.sqrt(X**2 + Y**2))

# Create contour plot
plt.contour(X, Y, Z, levels=20, cmap='viridis')

plt.colorbar(label='Intensity')

plt.title('Contour Plot of sin(sqrt(x² + y²))')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')



```python
#creating confussion metrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predict)
```

```python
#for creating / ploting contour plot
from matplotlib.colors import ListedColormap
X_point,y_point=X_train,y_train
X1, X2 = np.meshgrid(
    np.arange(start=X_point[:, 0].min() - 1, stop=X_point[:, 0].max() + 1, step=0.01),
    np.arange(start=X_point[:, 1].min() - 1, stop=X_point[:, 1].max() + 1, step=0.01)
)

plt.contourf(
    X1, X2,
    classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
    alpha=0.75,
    cmap=ListedColormap(('green', 'blue'))
)
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i, j in enumerate(np.unique(y_point)):
    plt.scatter(
        X_point[y_point == j, 0], X_point[y_point == j, 1],
        c=ListedColormap(('green', 'blue'))(i), label=j
    )

plt.title('K-NN Decision Boundary (Test Set)')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
